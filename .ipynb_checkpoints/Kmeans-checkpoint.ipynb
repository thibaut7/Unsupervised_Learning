```python
import pandas as pd
```

```python
#Charger et explorer les donnees
df = pd.read_csv('Live.csv')
df.head(10)
```

Les colonnes Column1, Column2, Column3 sont des colonnes qui ont des valeurs inexistentes. Nous allons les supprimer

```python
df  =  df.drop(['Column1', 'Column2', 'Column3', 'Column4'], axis = 1)
df.head(4)
```

Nous allons creer la fonction **duration** 


```python
from datetime import datetime
def duration(date):
    """
    Calcule le temps depuis la publication en seconde
    
    Args:
    date dans le format "%m/%d/%Y %H:%M"
    
    Returns:
    int :seconde depuis la publication jusqu'aujourd'hui
    
    """
    date_format = "%m/%d/%Y %H:%M"

# Convertir la chaîne de caractères en objet datetime
    date_obj = datetime.strptime(date, date_format)

# Obtenir le timestamp en secondes
    seconds = date_obj.timestamp()

    return seconds
```

Appliquons la fonction duration à la colonne status_published

```python
df['status_published'] = df['status_published'].apply(duration)
df.head(4)
```

Nous allons effectuer la classification sur en laissant les colonnes 1 et 2. Nous allons apres projeter la colonne **status_type** pour voir la correspondance avec la classification.
### Determinons le nombre de clusters
#### La silouhette

### Centrer et reduire les donnees

```python
import sklearn
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
Xcr = sc.fit_transform(df.iloc[:,2:])
Xcr
```

```python
# librairie pour évaluation des partitions
from sklearn import metrics
from sklearn import cluster
import matplotlib.pyplot as plt
import numpy as np
# utilisation de la métrique "silhouette"
# faire varier le nombre de clusters de 2 à 10
res = []
for k in range(2,10):
    km = cluster.KMeans(n_clusters=k).fit(Xcr)
    res.append(metrics.silhouette_score(Xcr,km.labels_))

# graphique
plt.title("Silhouette")
plt.xlabel("# of clusters")
plt.plot(np.arange(2,10,1),res)
plt.show()
```

### Elbow

```python
# Initialiser une liste vide pour stocker les dispersions intra-cluster
inertias = []

# faire varier le nombre de clusters de 1 à 10
for k in range(2, 11):
    # Effectuer le clustering K-means
    kmeans = cluster.KMeans(n_clusters=k).fit(Xcr)
    # Ajouter la dispersion intra-cluster à la liste
    inertias.append(kmeans.inertia_)

# Afficher le graphique du coude
plt.plot(range(2, 11), inertias, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Dispersion')
plt.title("Règle du coude")
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2).fit(Xcr)
kmeans.labels_
```

```python
df['label'] = kmeans.labels_
```
